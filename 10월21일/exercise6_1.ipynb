```python
import os
os.chdir('c:/PyStexam')

from matplotlib import font_manager, rc
font_path = "./data/malgun.ttf" 
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('./data/cctv_seoul.csv')
mycolors = sns.color_palette('hls',25)

df.set_index(['구별'], inplace=True)

plt.figure(figsize=(10,7))
df['CCTV수'].plot(kind='barh', grid=True, color=mycolors)
plt.title("시각화 과제(1) 각 구의 CCTV 설치 현황")
plt.ylabel('구이름')
plt.xlabel('각 구에 설치된 CCTV 댓수')
plt.savefig("output/hw1.png")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('./data/cctv_seoul.csv')
mycolors2 = sns.color_palette('PRGn',25) 

df.set_index(['구별'], inplace=True)
df.sort_values(by='CCTV수', ascending=False, inplace=True)

plt.figure(figsize=(10,5))
df['CCTV수'].plot(kind='bar', grid=True, color=mycolors2)
plt.title("시각화 과제(2) 각 구의 CCTV 설치 현황")
plt.xlabel('구이름')
plt.ylabel('각 구에 설치된 CCTV 댓수')
plt.xticks(rotation=30)
plt.savefig("output/hw2.png")

plt.show()
```
