```python
import os
os.chdir('c:/PyStexam')

from matplotlib import font_manager, rc
font_path = "./data/malgun.ttf" 
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('./data/cctv_seoul.csv')
df.set_index('구별', inplace=True)
```

```python
### 전처리
df['CCTV비율'] = df.CCTV수 / df.인구수 * 100
df['노인비율'] = df.고령자 / df.인구수 * 100
display(df.head())
```

```python
### 시각화
f, axes = plt.subplots(2, 1, figsize=(20, 13))

sns.barplot(x=df.index, y="CCTV비율", data=df, palette="deep", ax=axes[0])
axes[0].set_title("각 구의 인구수 대비 CCTV 비율",fontsize=17)
axes[0].set_xlabel(None)
axes[0].set_ylabel('CCTV비율',fontsize=13)

sns.lineplot(x=df.index, y="노인비율", data=df, color='orange',ax=axes[1],marker=True)
sns.scatterplot(x=df.index, y="노인비율", data=df, hue=df.index,
                legend=False, s=120, ax=axes[1])
axes[1].set_title("각 구의 인구수 대비 고령자 비율",fontsize=17)
axes[1].set_xlabel(None)
axes[1].set_ylabel('노인비율',fontsize=13)

plt.savefig('output/hw3.svg')
plt.show()
```
