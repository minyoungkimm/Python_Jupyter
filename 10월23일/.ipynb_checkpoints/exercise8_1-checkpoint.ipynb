```python
## 문제1
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
print('1-(1)')
x1 = iris_x.head(10).dropna(axis=0)
display(x1)
```

```python
print('1-(2)')
x2 = iris_x.dropna(axis=0, thresh=3)
display(x2.head(10))
```

```python
print('1-(3)')
x3 = iris_x.dropna(subset=['sepal_length','sepal_width'],axis=0)
display(x3.head(10))
```

```python
print('1-(4)')
x4 = iris_x.fillna(0)
display(x4.head(10))
```

```python
print('1-(5)')
mean = round(iris_x.mean(axis=0),1)
print(mean)
x5 = iris_x.fillna(mean)
display(x5.head(10))
```

```python
print('1-(6)')
x6 = iris_x.fillna(10)
display(x6.head(10))
```

```python
print('1-(7)')
x7 = iris_x.fillna(method='ffill')
display(x7.head(10))
```

```python
## 문제3
print('3-(1)')
click = pd.read_csv('data/product_click.log',sep=' ', header=None)
print(click.head())
print(click.info())
```

```python
print('3-(2)')
click.columns = ['oldclicktime','clickpid']
print(click.head())
```

```python
print('3-(3)')
click['oldclicktime'] = click['oldclicktime'].astype('object') 
print(click.info())
```

```python
print('3-(4)')
click['newclicktime'] = pd.to_datetime(click['oldclicktime'],format="%Y%m%d%H%M")
print(click.info())
print(click.head())
```

```python
print('3-(5)')
click['year'] = click['newclicktime'].dt.year
click['month'] = click['newclicktime'].dt.month
click['day'] = click['newclicktime'].dt.day
click['hour'] = click['newclicktime'].dt.hour
click['minute'] = click['newclicktime'].dt.minute
print(click.info())
display(click.head())
```
